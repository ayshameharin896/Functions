```python
#print the even numbers till n
def a(x):
  for i in range(1,x+1):
    if i%2==0:
      print(i)
a(int(input("enter a value")))
```

```python
#printing the odd numbers till n
def a(x):
  for i in range(1,x+1):
    if i%2!=0:
      print(i)
a(int(input("enter a value")))
```

```python
#finding the factorial of a number
def a(x):
  mul=1
  for i in range(1,x+1):
    mul*=i
    print(mul)
a(int(input("enter a value")))
```
